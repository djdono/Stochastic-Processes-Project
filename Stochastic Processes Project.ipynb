```python
data = pd.read_csv('Stock Data.csv')
dates = list(data.Date.unique())
stocks = list(data.Name.unique())

data.drop(['Open', 'High', 'Low', 'Volume'],axis=1, inplace = True)
df = pd.pivot_table(data, values= 'Close' , index = 'Name', columns = 'Date')
df.fillna(method='ffill', inplace = True)
df.fillna(method='ffill', inplace = True)
```

```python
S = []
for j in range(len(stocks)):
    S.append(0)

one_over_T = 1/(len(dates))
for i in range(len(dates)):
    a = np.array(df[dates[i]])
    s = one_over_T *(a*np.transpose(a))
    S = S+s
```

```python
S1 = []
for j in range(len(stocks)):
    S1.append(0)

one_over_T_minus_1 = 1/(len(dates)-1)
for i in range(len(dates)-1):
    a = np.array(df[dates[i]])
    b = np.array(df[dates[i+1]])
    s1 = one_over_T_minus_1  *(a*np.transpose(b))
    S1 = S1+s1
```

```python
m = []
for j in range(len(stocks)):
    m.append(1)
M= np.eye(len(stocks))

def summation(M):
    return sum(abs(M))
def constraint(M):
    return np.amax(abs((S*M)-np.transpose(S1)))
con = {'type':'ineq', 'fun':constraint}
nlc = NonlinearConstraint(constraint, 0, 1)
```

```python
A=minimize(summation, M, constraints = nlc).x
A
```
